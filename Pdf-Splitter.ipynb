```python
import PyPDF2
import tkinter as tk
from tkinter import filedialog

# Create the root window for the user interface
root = tk.Tk()
root.title("PDF Splitter")

# Create a function to handle the file selection using a file dialog
def select_file():
    file_types = [("PDF files", "*.pdf")]
    file_path = filedialog.askopenfilename(filetypes=file_types)
    if file_path:
        input_file_entry.delete(0, tk.END)
        input_file_entry.insert(0, file_path)
    else:
        tk.messagebox.showerror("Error", "Please select a PDF file.")

# Create a function to split the PDF into individual pages
def split_pdf():
    # Get the path to the input PDF file
    input_file_path = input_file_entry.get()
    
    # Open the PDF file in binary mode
    with open(input_file_path, 'rb') as infile:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(infile)
    
        # Get the total number of pages in the PDF document
        total_pages = len(pdf_reader.pages)
    
        # Create a PDF writer object
        pdf_writer = PyPDF2.PdfWriter()
        
        # Loop through each page in the PDF document
        for page_num in range(total_pages):
            # Get the current page from the PDF reader object
            page = pdf_reader.pages[page_num]
        
            # Determine if the current page is a left page or a right page
            if page_num % 2 == 0:
                # If the current page is a left page, add it to a new PDF document
                pdf_left = PyPDF2.PdfWriter()
                pdf_left.add_page(page)
            else:
                # If the current page is a right page, add it to the same PDF document as the left page
                pdf_left.add_page(page)
            
                # Add the combined left and right page PDF document to the output PDF writer object
                pdf_writer.add_page(pdf_left.pages[0])
                pdf_writer.add_page(pdf_left.pages[1])
        # Get the input file name without the extension
        input_file_name = input_file_path.split("/")[-1].split(".")[0]

        # Create the output file name by adding "mod" before the extension
        output_file_name = input_file_name + "_mod.pdf"
         
        # Write the output PDF document to a file
        with open(output_file_name, 'wb') as outfile:
            pdf_writer.write(outfile)
    # Display a message box to confirm the operation is complete
    tk.messagebox.showinfo("PDF Splitter", "PDF splitting complete.")

# Create the user interface elements
input_file_label = tk.Label(root, text="Input PDF file:")
input_file_entry = tk.Entry(root, width=80)
input_file_button = tk.Button(root, text="Select file", command=select_file)
split_button = tk.Button(root, text="Split PDF", command=split_pdf)

# Add the user interface elements to the root window
input_file_label.grid(row=0, column=0, padx=10, pady=20)
input_file_entry.grid(row=0, column=1, padx=10, pady=10)
input_file_button.grid(row=0, column=2, padx=10, pady=10)
split_button.grid(row=1, column=1, padx=10, pady=10)

# Run the user interface
root.mainloop()
```
